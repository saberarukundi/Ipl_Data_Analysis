#####  Total Matches Played by Each teams ?

```python
Teammatches = pd.concat([IPL_matches_df['Team1'],IPL_matches_df['Team2']]).value_counts()
Teammatches = pd.DataFrame(Teammatches).reset_index()
Teammatches.rename(columns={'index':'Teams','count':'Total_Matches'},inplace=True)
Teammatches
```

#### Visualization

> Matplotilb

```python
def get_color(name, number):
    pal = list(sns.color_palette(palette=name, n_colors=number).as_hex())
    return pal
```

```python
pal_vi = get_color('viridis_r', len(IPL_matches_df))
```

```python
pal_vi
```

```python
Teammatches.columns
```

```python
# Set figure size
plt.figure(figsize=(30, 10))

# Create bar plot
Teammatches1 = Teammatches.head(10)
bars = plt.bar(
    Teammatches1['Teams'],
    Teammatches1['Total_Matches'],
    color=pal_vi[:len(Teammatches1)]  # Use your color palette
)


plt.xlabel('Teams')
plt.ylabel('Total_Matches')
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height,
        f'{height}',
        ha='center',
        va='bottom',
        fontsize=20
    )

# Customize plot appearance
plt.xlabel('Teams')
plt.ylabel('Total Matches')
plt.xticks(rotation=45)
plt.gca().set_facecolor('white')   # Plot background
plt.gcf().set_facecolor('white')   # Figure background

# Remove top and right spines for a cleaner look
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Adjust layout
plt.tight_layout()
plt.show()
```

## Who is the best wicket taker bowler in IPL history?
